## Practical Lab 2 - Data Visualization and Publication

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Below is a scatter plot showing random data points.

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

#### Below is a bar chart displaying values for different categories.

```python
categories = ['Category A', 'Category B', 'Category C']
values = [20, 35, 45]

plt.bar(categories, values)
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

#### Below is a line plot of a sine wave.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

#### Below is a histogram showing the distribution of random data.

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
