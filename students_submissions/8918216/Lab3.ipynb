## Practical Lab 3 - Univariate Linear Regression on the Scikit-Learn Diabetes Dataset

#### Load the Dataset

```python
from sklearn import datasets
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

#### Explore the Data

```python
import matplotlib.pyplot as plt

# Plot BMI (independent variable) vs. Disease Progression (dependent variable)
plt.scatter(diabetes_X[:, 2], diabetes_y, c='b', marker='o', label='Data points')
plt.xlabel('BMI (Body Mass Index)')
plt.ylabel('Disease Progression')
plt.title('BMI vs. Disease Progression')
plt.legend()
plt.show()
```

#### Data Splitting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(diabetes_X[:, 2], diabetes_y, test_size=0.2, random_state=42)
```

#### Train a Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

# Create a linear regression model
model = LinearRegression()

# Fit the model on the training data
model.fit(X_train.reshape(-1, 1), y_train)
```

#### Report Results

```python
# Model parameters
slope = model.coef_[0]
intercept = model.intercept_

# Loss on train and test datasets
train_loss = model.score(X_train.reshape(-1, 1), y_train)
test_loss = model.score(X_test.reshape(-1, 1), y_test)

# Visualize the training and test data along with the model
plt.scatter(X_train, y_train, c='b', marker='o', label='Training data')
plt.scatter(X_test, y_test, c='g', marker='s', label='Test data')
plt.plot(X_train, model.predict(X_train.reshape(-1, 1)), c='r', label='Regression line')
plt.xlabel('BMI (Body Mass Index)')
plt.ylabel('Disease Progression')
plt.title('Linear Regression: BMI vs. Disease Progression')
plt.legend()
plt.show()

# Report results
print(f"Model Parameters: Slope = {slope:.2f}, Intercept = {intercept:.2f}")
print(f"Train Loss: {train_loss:.2f}")
print(f"Test Loss: {test_loss:.2f}")
```

## Takeaways:
#### - The intercept of the model is roughly 152.00, and the slope (coefficient) is approximately 998.58. These parameters suggest that starting from an intercept of 152.00, we can anticipate an increase in disease progression of about 998.58 units for every unit increase in BMI.<br><br>- The training and test datasets' R-squared values show how well the model fits the data. The R-squared value in this instance is 0.37 for the training data and 0.23 for the test data. The range of R-squared values is 0 to 1, with values closer to 1 indicating a better fit. The R-squared values in this situation imply that the linear regression model explains about 37% of the variability in the training data and 23% of the variability in the test data.<br><br>- The scatterplots of the training and test data points and the regression line visually illustrate the relationship between BMI and disease progression. The regression line captures a positive correlation between BMI and disease progression but does not fit the data points very tightly, especially in the test dataset.
